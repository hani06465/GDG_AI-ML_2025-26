```python
import numpy as np
```

```python
#  a list of five elements
lists = [6,4,7,8,19]
```

```python
# convert to nump list
lists = np.array(lists)
```

```python
#operation
Average= np.mean(lists)
Maximum = np.max(lists)
Total = np.sum(lists)
```

```python
#printing the results:
print("THE RESULTS ARE:")
print("Mean: ", Average)
print("Maximum: ", Maximum)
print("Sum: ", Total)
```
